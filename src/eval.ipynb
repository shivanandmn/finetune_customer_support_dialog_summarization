```python
from src.metrics import Metrics
import pandas as pd
import os
os.chdir("../")
os.getcwd()
```

```python
files = {
    "customer_care_dialog_summary_mistral_7b_v01": "results/customer_care_dialog_summary_mistral_7b_v01.csv",
    "customer_care_dialog_summary_llama2_7b": "results/customer_care_dialog_summary_llama2_7b.csv",
    "customer_care_dialog_summary_phi_2": "results/customer_care_dialog_summary_phi_2.csv",
    "mistral_7b_v01": "results/mistral_7b_v01.csv",
    "llama2_7b": "results/llama2_7b.csv",
    "phi_2": "results/phi_2.csv",
    
}
```

```python
import copy
df_merged = None
count = 0
for name, file in files.items():
    df = pd.read_csv(file)
    df = df.rename(columns={k: f'{k}_{name}' for k in df.columns if k!='summary'})
    if df_merged is None:
        df_merged = copy.deepcopy(df)
        
    else:
        df_merged = pd.merge(df_merged, df, how='inner', on='summary')
    count += df.shape[0]
print(f'Processed {count} rows')


df_merged.shape
```

```python
predicted_cols = [x for x in df_merged.columns if x.startswith("prediction")]
df_merged = df_merged[predicted_cols+["summary"]]
df_merged.isnull().sum()
```

```python
metrics = Metrics()
```

```python
results = {

}
for col in predicted_cols:
    score = metrics.score(prediction=df_merged[col], reference=df_merged["summary"])
    results[col] = score

print(results)
```

```python
import json

json.dump(results, open("results/eval_scores.json", "w"))
```
